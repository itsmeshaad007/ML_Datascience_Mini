```python
# By default it is inner join
pd.merge(data1, data2)
```

```python
pd.merge(data1, data2, how = 'left')
```

```python
data1
```

```python
# Everything is brought from both left and right table
pd.merge(data1, data2, how = 'outer')
```

```python
def class_p(Pclass):
    if Pclass == 1:
        return 'First'
    elif Pclass == 2:
        return 'Second'
    elif Pclass == 3:
        return 'Third'
    else:
        return 'NK'
```

```python
df['Pclass_str'] = df['Pclass'].apply(class_p)
```

```python
df.head(2)
```

```python
df['Pclass'].unique()
```

```python
def fare_seg(Fare):
    if Fare < 10:
        return 'Cheap'
    elif Fare > 10 and Fare < 20:
        return 'Mid'
    else:
        return 'High'
```

```python
df['flag_fare'] = df['Fare'].apply(fare_seg)
```

```python
df[['Fare', 'flag_fare']]
```
